```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
nifty_data = yf.download('^NSEI', start='2022-01-01', end='2024-12-31')
def calculate_bollinger_bands(data, window=20, num_std=2):
    close = data['Close']
    ma = close.rolling(window=window).mean()
    std = close.rolling(window=window).std()
    upper_band = ma + (num_std * std)
    lower_band = ma - (num_std * std)
    data['MA'] = ma
    data['Upper'] = upper_band
    data['Lower'] = lower_band
    return data

def calculate_macd(data, fast=12, slow=26, signal=9):
    data['EMA_Fast'] = data['Close'].ewm(span=fast, adjust=False).mean()
    data['EMA_Slow'] = data['Close'].ewm(span=slow, adjust=False).mean()
    data['MACD'] = data['EMA_Fast'] - data['EMA_Slow']
    data['Signal'] = data['MACD'].ewm(span=signal, adjust=False).mean()
    return data

def calculate_stochastic_oscillator(data, window=14, smooth_k=3):
    low_min = data['Low'].rolling(window=window).min()
    high_max = data['High'].rolling(window=window).max()
    data['%K'] = 100 * ((data['Close'] - low_min) / (high_max - low_min))
    data['%D'] = data['%K'].rolling(window=smooth_k).mean()
    return data
    
def calculate_vwap(data): #THIS IS INCORRECT AS VWAP SHOULD BE INTRADAY
    typical_price = (data['High'] + data['Low'] + data['Close']) / 3
    tp_volume = typical_price * data['Volume']
    cumulative_tp_volume = tp_volume.cumsum()
    cumulative_volume = data['Volume'].cumsum() 
    vwap = cumulative_tp_volume / cumulative_volume
    data['VWAP'] = vwap
    return data

def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    data['RSI'] = 100 - (100 / (1 + rs))
    return data

def calculate_atr(data, window=14):
    high_low = data['High'] - data['Low']
    high_close = np.abs(data['High'] - data['Close'].shift())
    low_close = np.abs(data['Low'] - data['Close'].shift())
    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = np.max(ranges, axis=1)
    data['ATR'] = true_range.rolling(window=window).mean()
    return data

nifty_data = calculate_bollinger_bands(nifty_data)
nifty_data = calculate_macd(nifty_data)
nifty_data = calculate_stochastic_oscillator(nifty_data)
nifty_data = calculate_vwap(nifty_data)
nifty_data = calculate_rsi(nifty_data)
nifty_data = calculate_atr(nifty_data)

def plot_bollinger_bands(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data.index, data['Close'], label='Close Price', color='blue')
    plt.plot(data.index, data['MA'], label='Moving Average', color='orange')
    plt.plot(data.index, data['Upper'], label='Upper Band', color='green', linestyle='--')
    plt.plot(data.index, data['Lower'], label='Lower Band', color='red', linestyle='--')
    plt.fill_between(data.index, data['Upper'], data['Lower'], color='grey', alpha=0.1)
    plt.title('Bollinger Bands')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_macd(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data['MACD'], label='MACD', color='blue')
    plt.plot(data['Signal'], label='Signal Line', color='red')
    plt.title('MACD')
    plt.legend()
    plt.show()

def plot_stochastic_oscillator(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data['%K'], label='%K', color='blue')
    plt.plot(data['%D'], label='%D (Signal)', color='red')
    plt.axhline(80, color='grey', linestyle='--')
    plt.axhline(20, color='grey', linestyle='--')
    plt.title('Stochastic Oscillator')
    plt.legend()
    plt.show()

def plot_vwap(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.plot(data['VWAP'], label='VWAP', color='orange')
    plt.title('Volume Weighted Average Price (VWAP)')
    plt.legend()
    plt.show()

def plot_rsi(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data['RSI'], label='RSI', color='blue')
    plt.axhline(70, color='red', linestyle='--')
    plt.axhline(30, color='green', linestyle='--')
    plt.title('Relative Strength Index (RSI)')
    plt.legend()
    plt.show()

def plot_atr(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data['ATR'], label='ATR', color='blue')
    plt.title('Average True Range (ATR)')
    plt.legend()
    plt.show()

# === Plot All Indicators ===
plot_bollinger_bands(nifty_data)
plot_macd(nifty_data)
plot_stochastic_oscillator(nifty_data)
plot_vwap(nifty_data)
plot_rsi(nifty_data)
plot_atr(nifty_data)
```
